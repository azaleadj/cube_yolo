```python
from ultralytics import YOLO
```

```python
# Load a pretrained model
model = YOLO("yolov8s.pt")
```

```python
model.train(
    # data="Rubix Cube.v3i.yolov8/data.yaml", 
    data="data.yaml", 
    epochs=50, 
    imgsz=640
)
```

```python
metrics = model.val()
print(metrics)
```
